Data Visualization

1. Scatter Plot without Key Attribute

```python
plt.figure(figsize = (15, 5))
plt.subplot(131)
sns.scatterplot(x="Age", y="Fare", data=data)
```

2. Scatter Plot with Key Attribute

```python
plt.figure(figsize = (15, 5))
plt.subplot(131)
sns.scatterplot(x="Age", y="Fare",hue="Sex", data=data)
```

2. Joint Plot :) Testing

```python
sns.jointplot(x="Age", y="Fare", hue="Sex", data=data)
```

Map

```python
plt.figure(figsize = (15,7))
sns.heatmap(data.corr(),annot=True,fmt='.1g',cmap='PRGn')
```

Box Plot

```python
sns.boxplot(x="Age", y="Sex", data=data)
```

Histogram

```python
plt.figure(figsize=(7,7))
sns.histplot(data["Age"], color='black')
plt.title('Distribution of {}'.format(5))
plt.show()
```

Line Chart

```python
data = data.head(50)
plt.plot(data['Age'],marker='o')
plt.plot(data['Fare'],marker='o')
plt.title ("Line Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

Stacked Bar Chart

```python
pd.crosstab(data['Pclass'], data['Sex']).plot(kind='bar', stacked='True')
plt.title('Stacked Bar Chart')
```
